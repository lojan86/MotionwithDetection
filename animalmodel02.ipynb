```python
import os

label_dir = "/content/drive/MyDrive/Colab Notebooks/animal_dataset/labels/train"
counts = {}

for file in os.listdir(label_dir):
    with open(os.path.join(label_dir, file)) as f:
        for line in f:
            cls = line.split()[0]
            counts[cls] = counts.get(cls, 0) + 1

print(counts)
```
